## Dummy Variables


A dummy variable is a numerical variable used in regression analysis to represent subgroups of the sample in your study. In research design, a dummy variable is often used to distinguish different treatment groups. In the simplest case, we would use a 0,1 dummy variable where a person is given a value of 0 if they are in the control group or a 1 if they are in the treated group. Dummy variables are useful because they enable us to use a single regression equation to represent multiple groups.


```python
import pandas as pd
import numpy as np
```

```python
sales=pd.read_csv("C:/Users/admin/IBM_Training/Training/Numpy - Series, DataFrame, Missing, Group, Merge/Sales_Outliers.csv")
```

```python
#The dataset contains few character categorical varaibles like "Outlet_Identifier", "Item_Fat_Content", "Outlet_Size", etc.
```

```python
# Working on Dummy_Variable

#x1=pd.get_dummies(sales["Item_Fat_Content"])
x1=pd.get_dummies(sales["Item_Fat_Content"],drop_first=True)
x1.head()
x2=pd.get_dummies(sales["Item_Type"],drop_first=True)
x2.head()
x3=pd.get_dummies(sales["Outlet_Identifier"],drop_first=True)
x3.head()
x4=pd.get_dummies(sales["Outlet_Size"],drop_first=True)
x4.head()
x5=pd.get_dummies(sales["Outlet_Location_Type"],drop_first=True)
x5.head()
x6=pd.get_dummies(sales["Outlet_Type"],drop_first=True)
x6.head()
```

```python
#concatenating the dummy variables into the original dataset

sales=pd.concat([sales,x1,x2,x3,x4,x5,x6],axis=1)
sales
```
